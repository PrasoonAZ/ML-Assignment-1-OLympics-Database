```python
team_participants_medals = df_olympics.groupby(['Year', 'Team'])['Name'].count().unstack(fill_value=0)
team_medals_each_year = medalists.groupby(['Year', 'Team'])['Medal'].count().unstack(fill_value=0)

most_participants_each_year = team_participants_medals.idxmax(axis=1)
most_medals_each_year = team_medals_each_year.idxmax(axis=1)

print("Teams with the most participants each year:")
print(most_participants_each_year)

print("\nTeams with the most medals each year:")
print(most_medals_each_year)
```

```python
medals_1896_2024 = df_olympics[df_olympics['Medal'] != 'No medal']

medal_counts = medals_1896_2024.groupby(['Team', 'Sport', 'Medal']).size().unstack(fill_value=0).reset_index()

medal_counts = medal_counts[['Team', 'Sport', 'Gold', 'Silver', 'Bronze']].fillna(0)

medal_counts['Total_Medals'] = medal_counts['Gold'] + medal_counts['Silver'] + medal_counts['Bronze']
most_medals_by_sport = medal_counts.loc[medal_counts.groupby('Sport')['Total_Medals'].idxmax()]

for index, row in most_medals_by_sport.iterrows():
    sport = row['Sport']
    team = row['Team']
    gold = int(row['Gold'])
    silver = int(row['Silver'])
    bronze = int(row['Bronze'])
    print(f"Sport: {sport}")
    print(f"Team with the Best Performance: {team}")
    print(f"Gold Medals: {gold}")
    print(f"Silver Medals: {silver}")
    print(f"Bronze Medals: {bronze}\n")
```

Multivariate Analysis

```python
'''
This analysis focuses on visualizing the total medals won by the top 10 countries in the Olympic Games over the years, using a heatmap
for clear representation.

First, we filter the dataset to exclude entries where athletes did not win any medals. We then group the data by 'Team' and count
the total number of medals for each country.

Next, we identify the top 10 teams based on their total medal counts. Using this subset, we further group the data by 'Team' and
'Year', counting the number of medals won by each country in each Olympic year.

We then create a heatmap to visualize this data. The heatmap displays medal counts for each of the top 10 countries across various years,
with color intensity indicating the number of medals won.
This visualization provides insights into how the performance of these countries has evolved over time, highlighting trends
in Olympic success and participation.
'''

# Group by Team and count total medals, excluding 'No Medal'
total_medals = df_olympics[df_olympics['Medal'] != 'No medal'].groupby('Team').size()

top_10_teams = total_medals.nlargest(10).index

medals_by_country_year = df_olympics[df_olympics['Team'].isin(top_10_teams) & (df_olympics['Medal'] != 'No medal')].groupby(['Team', 'Year']).size().unstack(fill_value=0)

plt.figure(figsize=(15, 8))
sns.heatmap(medals_by_country_year, cmap='YlGnBu', linewidths=0.5)
plt.title('Heatmap of Medals Won by Top 10 Countries and Year')
plt.xlabel('Year')
plt.ylabel('Country')
plt.show()
```

```python
'''
Performing multivariate analysis to visualize the distribution of Olympic medals among the top 10 NOCs
and top 10 sports using a heatmap.

First, we identify the top 10 countries by counting the occurrences of NOC and the top 10 sports based on
participation.

Next, we filter the dataset to include only those entries that belong to the selected top countries and sports, while excluding
athletes who did not win any medals.

We then group the filtered data by NOC, Sport, and Medal type, counting the occurrences of each combination. This results in a pivot
table where rows represent NOCs, columns represent medal types, and values indicate the count of medals won.

Finally, we plot a heatmap to visualize this data. The heatmap displays the counts of medals won by each country in each sport,
with color intensity representing the number of medals.
This visualization provides insights into which countries excel in specific sports and how many medals they have won,
offering a comprehensive view of Olympic performance among the top competitors.
'''


top_noc = df_olympics['NOC'].value_counts().nlargest(10).index
top_sports = df_olympics['Sport'].value_counts().nlargest(10).index

medals_by_team_sport_type = df_olympics[df_olympics['NOC'].isin(top_noc) & df_olympics['Sport'].isin(top_sports) & (df_olympics['Medal'] != 'No medal')]
medals_count_noc = medals_by_team_sport_type.groupby(['NOC', 'Sport', 'Medal']).size().unstack(fill_value=0)

# Plot the heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(medals_count_noc, cmap='YlGnBu', linewidths=0.5)
plt.title('Heatmap of Medal Counts by Team (NOC), Sport, and Medal Type (Top 10)')
plt.xlabel('Medal Type')
plt.ylabel('Country (NOC)')
plt.xticks(rotation=45)
plt.show()
```

```python
"""
Performing multivariate analysis by creating a line plot to visualize trends in medal counts over the years for various sports in the df_olympics dataset.

The 'medal_counts_by_year' variable groups the dataset by year, sport, and NOC to count the total number of medals won by each country in different sports for each Olympic year.
This allows us to analyze the progression of medal wins over time.

Using the seaborn `lineplot`, we display the number of medals won over the years for each sport, with lines differentiated by colors representing different sports.
Markers are used at each data point for clarity.

This visualization offers valuable insights into the changing trends in medal counts across sports over the years, highlighting periods of dominance in certain sports for specific countries.
It also helps in identifying potential shifts in which sports or countries have excelled in the Olympics historically.
"""
pair_data = df_olympics[
    df_olympics['NOC'].isin(top_noc) &  # Filter by top NOCs
    df_olympics['Sport'].isin(top_sports) &  # Filter by top sports
    (df_olympics['Medal'] != 'No medal')  # Exclude rows with 'No medal'
]

medal_counts_by_year = pair_data.groupby(['Year', 'Sport', 'NOC']).size().reset_index(name='Medal_Count')

plt.figure(figsize=(10, 6))
sns.lineplot(data=medal_counts_by_year, x='Year', y='Medal_Count', hue='Sport', marker='o', palette='Set2')
plt.title('Medal Count Over Years by Sport')
plt.ylabel('Number of Medals')
plt.xlabel('Year')
plt.xticks(rotation=45)
plt.legend(title='Sport')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
'''
Performing multivariate analysis to explore the distribution of Olympic medals across different sports and countries (NOCs)
in the df_olympics dataset.

A new column, 'Medal_Type', is created to categorize the types of medals (Gold, Silver, Bronze) for easier plotting.

The FacetGrid is then set up to create individual subplots for each combination of sport and country (NOC).
Each subplot displays a count plot showing the distribution of medal types for that specific sport and country.

This visualization provides insights into how different countries perform in various sports, highlighting the types of
medals they have won.
By examining these facets, we can identify trends in performance across sports and countries, offering a comprehensive
view of Olympic achievements.

'''

facet_data = df_olympics[df_olympics['NOC'].isin(top_noc) & df_olympics['Sport'].isin(top_sports) & (df_olympics['Medal'] != 'No medal')]
facet_data['Medal_Type'] = facet_data['Medal']

g = sns.FacetGrid(facet_data, row='Sport', col='NOC', margin_titles=True, height=4)
g.map_dataframe(sns.countplot, x='Medal_Type', palette='Set2')
g.set_axis_labels("Medal Type", "Count")
g.set_titles(col_template="{col_name}", row_template="{row_name}")
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Medal Distribution by Sport, Country (NOC), and Medal Type')
plt.show()
```

```python
'''
Performing analysis to identify the top 10 athletes with the most Olympic medals in the df_olympics dataset.
The variable 'top_athletes' filters out entries where athletes did not win a medal and groups the data by athlete names, counting the
total number of medals each athlete has won.

Next, we retrieve the countries associated with these top athletes by filtering the dataset for their names and extracting unique entries
for 'Name' and 'Team'.  This allows us to understand which countries these successful athletes represent.

We then merge the medal counts with their respective countries to create a comprehensive DataFrame containing each athlete's name,
medal count, and country.

To enhance visualization, we define distinct colors for each country represented in the top athletes list. This helps in visually
distinguishing between athletes from different countries in the subsequent bar chart.

Finally, we plot a bar chart that displays the top 10 athletes and their medal counts, using color coding to indicate their
respective countries. This analysis provides insights into which nations have produced the most successful athletes in Olympic history
and highlights their achievements based on medal counts.
'''

# Top 10 athletes winning the most medals
top_athletes = df_olympics[df_olympics['Medal'] != 'No medal'].groupby('Name')['Medal'].count().sort_values(ascending=False).head(10)

# Get the countries for the top athletes
top_athlete_names = top_athletes.index
top_athlete_countries = df_olympics[df_olympics['Name'].isin(top_athlete_names)][['Name', 'Team']].drop_duplicates()

# Merge the medal counts with their respective countries
top_athletes_df = top_athletes.reset_index().merge(top_athlete_countries, on='Name')

# Define a color for each country
unique_countries = top_athletes_df['Team'].unique()
colors = sns.color_palette("hsv", len(unique_countries))  # Generate distinct colors
country_color_map = {country: colors[i] for i, country in enumerate(unique_countries)}

# Map the colors to each athlete's country
top_athletes_df['Color'] = top_athletes_df['Team'].map(country_color_map)

# Plotting the bar chart with only athlete names as x labels
plt.figure(figsize=(12, 6))
bars = plt.bar(top_athletes_df['Name'], top_athletes_df['Medal'], color=top_athletes_df['Color'])
plt.title('Top 10 Athletes Winning the Most Medals')
plt.ylabel('Medal Count')
plt.xticks(rotation=45)

# Create a custom legend
handles = [plt.Rectangle((0, 0), 1, 1, color=country_color_map[country]) for country in unique_countries]
plt.legend(handles, unique_countries, title='Country', loc='upper right')

plt.show()

```
